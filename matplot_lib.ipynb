```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = [1,2,3,4]
```

```python
# Create a subplot to investigate the properties of 'Axes'
ncols, nrows = (3,3)
fig, axes = plt.subplots(ncols, nrows, figsize=(nrows * 3, ncols * 3))
axes[0,0].plot(t)
```

```python
print("type of axes: " + str(type(axes)))
print("shape of axes: " + str(axes.shape))
```
